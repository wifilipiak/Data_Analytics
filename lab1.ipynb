```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import csv file and set column 1 as the index
df = pd.read_csv ('Data1.csv', index_col = 0)
df.head()
```

```python
#convert index to datetime
df.index = pd.to_datetime(df.index)
df
```

```python
# plot all columns as timeseries
plt.plot(df.index,df)
plt.legend(["Theta 1", "Theta 2", "Theta 3", "Theta 4", "Theta 5", "Theta 6"])
plt.show()
```

```python
melted = df.melt()
melted.head()
```

```python
# plot histograms for all variables
a = sns.FacetGrid(melted, col="variable", col_wrap=3, sharex=False)
a = a.map(plt.hist, "value", bins = 10)
```

```python
# plot Kernel Denisty Estimators for all columns
a = sns.FacetGrid(melted, col="variable", col_wrap=3, sharex=False)
a = a.map(sns.kdeplot, "value")
```

```python
# select thetas 1-4 in year 2018
df_filtered = df.loc['2018-01-01':'2018-12-31','theta_1':'theta_4']
df_filtered.head()
```

```python
# plot all columns as timeseries
plt.plot(df_filtered.index,df_filtered)
plt.legend(["Theta 1", "Theta 2", "Theta 3", "Theta 4"])
plt.show()
```

```python
melted_fil = df_filtered.melt()
# plot histograms for all variables
c = sns.FacetGrid(melted_fil, col="variable", col_wrap=2, sharex=False)
c = c.map(plt.hist, "value", bins = 10)
```
