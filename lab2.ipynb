```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.special import expit
```

```python
#creating normal distributions with 10000 samples and different mean and variance parameters  mean - M , variance - V
#(M = 0, V = 0.1), (M = 10, V = 100), (M = 0, V = 1), (M = 50 , V = 25)
data_norm = {'M,V (0, 0.1)': np.random.normal(0, 0.1, 10000), 'M,V (10, 100)':np.random.normal(10, 10, 10000), 'M,V (0, 1)':np.random.normal(0, 1, 10000),'M,V (50, 25)':np.random.normal(50, 5, 10000) }
df_norm = pd.DataFrame(data = data_norm)
#creating inverse logit functions of created earlier normal distributions
data_log = {'M,V (0, 0.1)': expit(np.random.normal(0, 0.1, 10000)), 'M,V (10, 100)':expit(np.random.normal(10, 10, 10000)), 'M,V (0, 1)':expit(np.random.normal(0, 1, 10000)),'M,V (50, 25)':expit(np.random.normal(50, 5, 10000))}
df_log = pd.DataFrame(data = data_log)
#creating Poisson distributions with 10000 samples and different lambda parameters  lambda - L
#(L = 0.1) , (L = 1), (L = 5), (L = 50)
data_pois = {'L (0.1)': np.random.poisson(0.1, 10000), 'L (1)': np.random.poisson(1, 10000), 'L (5)': np.random.poisson(5, 10000),'L (50)': np.random.poisson(50, 10000) }
df_pois = pd.DataFrame(data = data_pois)
#creating beta distributions with 10000 samples and different alpha and beta parameters A - alpha, B - beta
#(A = 1, B = 1), (A = 1, B = 100), (A = 100, B = 1), (A = 10, B = 100), 
data_beta = {'A,B (1, 1)': np.random.beta(1, 1, 10000), 'A,B (1, 100)': np.random.beta(1, 100, 10000), 'A,B (100, 1)': np.random.beta(10, 1, 10000),'A,B (10, 100)': np.random.beta(10, 100, 10000)}
df_beta = pd.DataFrame(data = data_beta)
#creating log-normal distributions with 10000 samples and different mean and variance parameters  mean - M , variance - V
#(M = 0, V = 0.1), (M = 10, V = 100), (M = 0, V = 1), (M = 50 , V = 25)
data_log_norm = {'M,V (0, 0.1)': np.random.lognormal(0, 0.1, 10000), 'M,V (10, 100)':np.random.lognormal(10, 10, 10000), 'M,V (0, 1)':np.random.lognormal(0, 1, 10000),'M,V (50, 25)':np.random.lognormal(50, 5, 10000) }
df_log_norm = pd.DataFrame(data = data_norm)
```

```python
def plot_non_cumulative(df):
    melted = pd.melt(df)
    g1 = sns.FacetGrid(melted, col = "variable", sharex = True, sharey = False)
    g1 = g1.map(plt.hist, 'value', density = True, cumulative = False, bins = 20)
```

```python
def plot_cumulative(df):
    melted = pd.melt(df)
    g2 = sns.FacetGrid(melted, col = "variable", sharex = True, sharey = False)
    g2 = g2.map(plt.hist, 'value', density = True, cumulative = True, bins = 20)
```
