```python
multiNormal3d = """
data {
    vector[3] Mu;
    matrix[3, 3] Sigma;
}
generated quantities {
    vector[3] y = multi_normal_rng(Mu, Sigma);
}
"""
multiNormal3d_model = pystan.StanModel(model_code = multiNormal3d)
```

```python
dataMultiNormal3d = {'Mu': np.array([0, 0, 0]), 'Sigma': np.array([[1, 0.5, 0.25], [0.5, 1, 0.75], [0.25, 0.75, 1]])}
multiNormal3d_sam = multiNormal3d_model.sampling(data = dataMultiNormal3d, algorithm = 'Fixed_param', iter = 1000)
multiNormal3d_df = pd.DataFrame(multiNormal3d_sam.extract()['y'])
multiNormal3d_df.columns = ['y1', 'y2', 'y3']
```

```python
plt.figure(figsize = [20, 20])
plt.subplot(1, 3, 1).set_title('y1')
plt.hist(multiNormal3d_df['y1'], bins = 50)
plt.grid()
plt.subplot(1, 3, 2).set_title('y2')
plt.hist(multiNormal3d_df['y2'], bins = 50)
plt.grid()
plt.subplot(1, 3, 3).set_title('y3')
plt.hist(multiNormal3d_df['y3'], bins = 50)
plt.grid()
plt.show()
```
